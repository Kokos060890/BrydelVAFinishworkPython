2.1. Изучите распределение домов от наличия вида на набережную
Постройте график

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузите данные из CSV файла в DataFrame
df = pd.read_csv('kc_house_data.csv')

# Постройте график распределения домов от наличия вида на набережную
plt.figure(figsize=(20, 5))
plt.hits=(df['waterfront'].value_counts().plot(kind='bar', color='blue'))
plt.title('Распределение домов от наличия вида на набережную')
plt.xlabel('Наличие вида на набережную')
plt.ylabel('Количество домов')
```

2.2. Изучите распределение этажей домов

```python
# Постройте гистограмму распределения этажей домов
plt.figure(figsize=(8, 6))
plt.hist(df['floors'], bins=10, color='green', alpha=0.7)
plt.title('Распределение этажей домов')
plt.xlabel('Этажи')
plt.ylabel('Количество домов')
plt.show()
```

2.3. Изучите распределение состояния домов

```python
# Постройте гистограмму распределения состояния домов
plt.figure(figsize=(8, 6))
plt.hist(df['yr_renovated'], bins=10, color='orange')
plt.title('Распределение состояния домов')
plt.xlabel('yr_renovated')
plt.ylabel('Количество домов')
plt.xticks(rotation=0)

```

#Общий вывод: Большинство домов в данном наборе данных не были реновированы, и информация о годе реновации отсутствует у многих домов. Также, небольшое количество домов было реновировано в определенные годы, что может указывать на периоды активной реновации в определенные временные отрезки

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


Диаграмма 1: График зависимости стоимости недвижимости от площади

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['sqft_living'], df['price'], alpha=0.5)
plt.title('Зависимость стоимости недвижимости от площади')
plt.xlabel('Площадь (кв. футы)')
plt.ylabel('Стоимость')
plt.show()
```

Вывод 1: На графике мы видим положительную зависимость между площадью недвижимости и ее стоимостью. С увеличением площади недвижимости обычно увеличивается и ее стоимость.

Диаграмма 2: График зависимости стоимости недвижимости от количества спален

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df.groupby('bedrooms')['price'].median())
plt.title('Зависимость стоимости недвижимости от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Стоимость')
plt.xticks(rotation=0)
plt.show()
```

Вывод 2: Из графика видно, что стоимость недвижимости имеет различия в зависимости от количества спален. Дома с большим количеством спален могут иметь более высокую стоимость.

Диаграмма 3: График зависимости стоимости недвижимости от года постройки

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['yr_built'], df['price'], alpha=0.5)
plt.title('Зависимость стоимости недвижимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость')
plt.show()
```

Вывод 3: График показывает, что старые дома не всегда имеют низкую стоимость, и наоборот, недвижимость, построенная в более поздние годы, также может иметь различные уровни стоимости.

Диаграмма 4: График зависимости стоимости недвижимости от наличия вида на набережную

```python
plt.figure(figsize=(8, 6))
df.groupby('waterfront')['price'].median().plot(kind='bar', color='blue')
plt.title('Зависимость стоимости недвижимости от наличия вида на набережную')
plt.xlabel('Наличие вида на набережную')
plt.ylabel('Стоимость')
plt.xticks(rotation=0)
plt.show()
```

Вывод 4: Дома с видом на набережную, вероятно, будут иметь более высокую стоимость по сравнению с домами без такого вида.

Диаграмма 5: График зависимости стоимости недвижимости от состояния

```python
plt.figure(figsize=(8, 6))
df.groupby('condition')['price'].median().plot(kind='bar', color='green')
plt.title('Зависимость стоимости недвижимости от состояния')
plt.xlabel('Состояние')
plt.ylabel('Стоимость')
plt.xticks(rotation=0)
plt.show()
```

Вывод 5: Состояние дома может влиять на его стоимость. Дома в хорошем состоянии, возможно, будут иметь более высокую стоимость.

Общий вывод:Каждая из представленных диаграмм и выводов помогает нам понять, как различные характеристики недвижимости могут влиять на ее стоимость. Этот анализ может помочь потенциальным покупателям и продавцам принимать более основательные решения.
